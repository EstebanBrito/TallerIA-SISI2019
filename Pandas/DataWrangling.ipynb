# Unión (Merging)

```python
# NOS PERMITE COMBINAR LA INFORMACIÓN DE DOS TABLAS EN UNA
df1 = pd.DataFrame({'key': ['b','b','a','c','a','a','b'],
                   'data1': range(7)})
df2 = pd.DataFrame({'key': ['a','b','d'],
                   'data2':range(3)})
df1
df2
```

## Default

```python
# Unión Many-to-many
pd.merge(df1, df2) # Se necesitan filas en común
```

```python
# Especifica la columna llave (se toman columnas con igual nombre por defecto)
# Es una buena práctica definir la llave
pd.merge(df1, df2, on='key')
```

```python
# Si los nombres de las columnas son diferentes, especifícalas (Unión L-R)
df3 = pd.DataFrame({'lkey': ['b','b','a','c','a','a','b'],
                   'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a','b','d'],
                   'data2': range(3)})
df3
df4
# Nota: Sigue siendo una unión tipo Inner
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
# Usemos una unión tipo Outer (se incluyen los índices no comunes c y d)
df1
df2
pd.merge(df1, df2, how='outer') # c y d aparecen como NaN
```

#### Tipos de unión (join)
inner: Por defecto. Sólo se incluyen combinaciones de llaves encontradas en las dos tablas
outer: Se incluen todas las llaves encontradas en ambas tablas
left, right: Se usan todas la sllaves encontradas en la tabla izquierda o derecha

```python
## MÁS SOBRE UNIONES
df1 = pd.DataFrame({'key': ['b','b','a','c','a','b'],
                   'data1': range(6)})
df2 = pd.DataFrame({'key': ['a','b','a','b','d'],
                   'data2':range(5)})
df1
df2
```

```python
# Uniones Many-to-many joins usan el producto cartesiano
# 3 filas-b en df1, 2 en df2: 3x2 = 6 filas-b en total
pd.merge(df1, df2, on='key', how='left')
```

```python
# UNIONES RIGHT, LEFT
pd.merge(df1, df2, how='inner') # c y d no aparecen
```

```python
pd.merge(df1, df2, how='left') # c aparece pero d no
```

```python
pd.merge(df1, df2, how='right') # d aparece pero c no
```

```python
## UNIONES POR MÚLTIPLES LLAVES
# Se utilizan las combinaciones de valores como llaves 
left = pd.DataFrame({'key1': ['foo','foo','bar'],
                    'key2': ['one','two','one'],
                    'lval': [1,2,3]})
right = pd.DataFrame({'key1': ['foo','foo','bar','bar'],
                    'key2': ['one','one','one','two'],
                    'rval': [4,5,6,7]})
left
right
pd.merge(left, right, on=['key1', 'key2'], how='outer')

# Notice the 1x2 in [foo,one] and the NAN values in [foo,two] and [bar,two]
```

```python
# Tratamiento de columnas superpuestas (sufijos por defecto)
pd.merge(left, right, on='key1')
# Sufijos personalizados (se identifica el origen de las columnas fácilmente)
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

## Utilizando índices para unir

```python
left1 = pd.DataFrame({'key': ['a','b','a','a','b','c'],
                     'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a','b'])
left1
right1
```

```python
# right_index=True es similar a right_on=right1.index
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
# Union tipo Outer.
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

Nota: Y todavía podemos hacer uniones con DataFrames con índices jerárquicos

# Concatenación

```python
# NOS PERMITE ACOMPLETAR DFs O SERIES
# CONCATENACIÓN DE SERIES
# 3 Series sin superposición de índices
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
```

```python
# Concatenamos a lo largo del índice y de las columnas
# Nótese los valores nulos escalonados
pd.concat([s1, s2, s3]) #->Series
pd.concat([s1, s2, s3], axis=1, sort=False) #->DF
```

```python
# Optamos por una unión Inner, más pequeña
s4 = pd.concat([s1, s3]) #->Series
s4
pd.concat([s1, s4], axis=1, join='inner') # f y g no aparecen
```

```python
# Podemos darles nombres a las columnas del DF concatenado
pd.concat([s1, s2, s3], axis=1, keys=['one','two','three'], sort=False)
```

```python
# CONCATENACIÓN CON DATAFRAMES
df1 = pd.DataFrame(np.arange(6).reshape(3,2), index=['a','b','c'],
                  columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2,2), index=['a','c'],
                  columns=['three', 'four'])
df1
df2
```

```python
# Al personalizar el nombre de las columnas, creamos un IJ
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'], sort=False)
# Otra manera de hacerlo: Usar diccionarios
pd.concat({'level1': df1, 'level2': df2}, axis=1, sort=False)
```

```python
# ATENCIÓN: Los índices numéricos son ignorados al concatenar
df1 = pd.DataFrame(np.random.randn(3,4), columns=['a','b','c','d'])
df2 = pd.DataFrame(np.random.randn(2,3), columns=['b','d','a'])
df1
df2
pd.concat([df1, df2], ignore_index=True, sort=False)
```

# Datos con superposición (Patching)

```python
# APLICAMOS UN DF SOBRE OTROS DFs PARA RELLENAR NAN Y OBTENER UN DF MÁS COMPLETO
# PATCHING CON SERIES
# Dos Series cuyos índices se sobreponen parcial o totalmente
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
             index=['f', 'e', 'd', 'c', 'b', 'a'])
b = pd.Series(np.arange(len(a), dtype=np.float64),
             index=['f', 'e', 'd', 'c', 'b', 'a'])
b[-1] = np.nan
a
b
```

```python
# RELLENAMOS LOS VALORES NAN EN UN DF USANDO LOS VALORES DE OTRO DF
a.combine_first(b)
# Originalmente hecho así:
np.where(pd.isnull(a), b, a)
```

```python
# PATCHING CON DATAFRAMES
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
                                  'b': [np.nan, 2., np.nan, 6.],
                                  'c': range(2, 18, 4)})
df2 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                   'b': [np.nan, 3., 4., 6., 8.]})
df1
df2
```

```python
df1.combine_first(df2)
```
